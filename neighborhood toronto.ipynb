```python
import requests
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python

My_table = soup.find('table',{'class':'wikitable sortable'})
My_table
```

```python
columns="Postcode,Borough,Neighbourhood"
table1=""
for tr in My_table.find_all('tr'):
    row1=""
    for tds in tr.find_all('td'):
        row1=row1+","+tds.text
    table1=table1+row1[1:]
file=open("toronto.csv","wb")
#file.write(bytes(headers,encoding="ascii",errors="ignore"))
file.write(bytes(table1,encoding="ascii",errors="ignore"))
```

```python
import pandas as pd
df = pd.read_csv('toronto.csv',header=None)
df.columns=["Postalcode","Borough","Neighbourhood"]
```

```python
df.head(10)
```

```python
df=df[df['Borough']!= 'Not assigned']
df.head(10)
```

```python
data=df.reset_index()
data.head()
```

```python
data=data[["Postalcode","Borough","Neighbourhood"]]
data
```

```python
data.loc[data['Neighbourhood'] =='Not assigned' , 'Neighbourhood'] = data['Borough']
result = data.groupby(['Postalcode','Borough'], sort=False).agg( ', '.join) 
result.head()
```

```python
result.shape
```

```python
result = result.reset_index()
result.head()
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="Toronto")
location1 = geolocator.geocode(address)
latitude_toronto = location1.latitude
longitude_toronto = location1.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude_toronto, longitude_toronto))

```

```python
map_toronto = folium.Map(location=[latitude_toronto, longitude_toronto], zoom_start=10)

# add markers to map
for lat, lng, borough, Neighbourhood in zip(Toronto['Latitude'], Toronto['Longitude'], Toronto['Borough'], Toronto['Neighbourhood']):
    label = '{}, {}'.format(Neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
